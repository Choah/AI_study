## RNN

CNN에는 분산 의존이 없지만 RNN은 알고리즘적으로 의존성이 존재
- 따라서 속도가 느리다. 
- CNN 보다 5 -13배 느리다 
* 장점: 문장을 생산하거나 하는거에 쉽게 할 수 있다. 

아웃풋같은 경우 길이를 무한정 생선할 수 있다.
단점으로는 정보가 굉장이 타임시퀀스가 길어질 수록 데이터를 잃어버리고 
gradient vanishing/ exploding으로 계산이 좀 불안정하다. 

따라서 이를 해결하기 위한 여러 기법들이 나오고 있다. 

약점을 보안하기 위해서 한방향으로 보내게 될 것을 input 문장을 2가지 버전으로 보낸다.

하나는 정방향으로 흐르도록 하고 하나는 문장 순서를 뒤집어서 역방향으로 가도록 한다. 

* Gradient vanishing/ exploding이 큰 이슈이다. 
-> LSTM 과 같은 기법이 이를 해결하기 위해 나온다.
        - forget gate: 이전 상태를 얼마나 잊을지
        - input gate: 현재 입력 종보를 얼마나 받을지
        - output gate: 새로 계산된 정보를 얼마나 상태정보로 출려할지 

-> 해결책: gradient가 어느 정도 커지면 그냥 잘리버리면 된다.
