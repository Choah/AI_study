```python
import json
```

```python
# dump/ dumps 
# Serialize 시키기 

name = ("Kim","Lee","Park")
age= [30,29,31]
person = {"name":"Kim", "age":12, "car":False}
```

```python
nameStr = json.dumps(name)
ageStr = json.dumps(age)
personStr = json.dumps(person, indent = "  ")
```

```python
# 아스키 코드  
kname = ("김","조","정")
knameStr = json.dumps(kname)
print(knameStr)

# 아스키 코드 무시
knameStr = json.dumps(kname, ensure_ascii=False)
print(knameStr)
```

```python
# load/loads
# parse와 비슷한 것 : string type -> object 

personObj = json.loads(personStr)
person == personObj
```

```python
# with a file 
# serialized 된 걸 object 로 나타낼 수 있다. 

person["car"] = ["레이","모닝"]

with open("person.json", "w") as f:
    json.dump(person,f)
    
with open("person.json", "r") as f:
    personObj=json.load(f)
    
personObj
```
