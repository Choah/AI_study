* HTTP Status Codes

- 1xx information: 100 - server agrees to handle client's request
- 2xx Success: 200 - request succeeded; 204 - no content present
- 3xx Redirection: 301 - page moved: 304 - cached page still valid
- 4xx Client error: 403 - forbidden page; 404 - page not found
- 5xx Server error: (일시적으로 서버가 멈추면) 500 - internal server error; 503 - try again later

```python
import requests
def download(method, url, param=None, data=None): # 없을 수도 있는 애들 None으로 초기값 준 것
    try: 
        resp = requests.request(method, url, params=param, data=data, headers=headers)
        resp.raise_for_status() # 에러나오면 알려주는 함수
    except requests.exceptions.HTTPError as e:
        if 500 <= e.response.status_code < 600:
            print("재시도")
        else:
            print(e.response.status_code)
            print(e.response.reason)
    return resp
```

```python
import requests
import time
def download(method, url, param=None, data=None, timeout=1, maxretries=3): # 없을 수도 있는 애들 None으로 초기값 준 것
    try: 
        resp = requests.request(method, url, params=param, data=data, headers=headers)
        resp.raise_for_status() # 에러나오면 알려주는 함수
    except requests.exceptions.HTTPError as e:
        if 500 <= e.response.status_code < 600 and maxretries>0:
            print(maxretries)
            time.sleep(timeout) # timeout 만큼 멈췄다가 뜨는 것 
            download(method, url, param, data, timeout, maxretries-1)
        else:
            print(e.response.status_code)
            print(e.response.reason)
    return resp
```
