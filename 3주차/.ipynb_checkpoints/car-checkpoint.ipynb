```python
from bs4 import BeautifulSoup
```

```python
import requests
import time

def download(method, url,
             param=None, data=None,
            timeout=1, maxretries=3):
    try:
        resp = requests.request(method, url,
                     params=param, data=data,
                               headers=headers)
        resp.raise_for_status()
    except requests.exceptions.HTTPError as e:
        if 500<= e.response.status_code <600\
        and maxretries >0:
            time.sleep(timeout)
            resp = download(method, url, param, data, timeout, maxretries-1)
            print("재시도")
        else:
            print(e.response.status_code)
            print(e.response.reason)
    return resp

headers = {"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36"}
```

```python
def parseURL(seed):
    
    html = download("get",seed)
    dom = BeautifulSoup(html.text, "lxml")
    
    return [requests.compat.urljoin(url, _.find_parent()["href"]) 
              for _ in dom.select(".fix .title") 
              if _.find_parent() and _.find_parent().has_attr("href")]
```

```python
url = "http://xn--sk-q81iz8ub6h66vkeav89a.com/product/"
html = download("get", url, param={"SM":"korea"})
dom = BeautifulSoup(html.text, "lxml")
```

```python
# a: 애들 다 가져오면 디텍팅해서 598까지 다 가져올 수 있을 것 

seed = list()
seed.extend([requests.compat.urljoin(url, _["href"])
            for _ in dom.select (".paging_inner a")])
seen = list()
queue= list()

while seed:
    baseURL = seed.pop(0)
    seen.append(baseURL)
    linkList = parseURL(baseURL)
    
    for link in linkList:
        if link not in queue and link not in seen:
            queue.append(link)
    print("Queue: {0}, Seen: {1}".format(len(queue), len(seen)))
```

```python
def parseContent(url):
    html = download("get", url)
    dom= BeautifulSoup(html.text, "lxml")
    
    return {"car": dom.select(".view_title h2")[0].text.strip(), # strip: 공백 제거 
            "price": dom.select(".car_price")[0].text.strip(),
            "model_yr": dom.select(".info_table td")[1].text.strip(),
            "fuel_efficiency": dom.select(".info_table td")[3].text.strip(),
            "drive": dom.select(".info_table td")[5].text.strip(),
            "engine_diaplacement": dom.select(".info_table td")[7].text.strip(),
            "shift": dom.select(".info_table td")[9].text.strip(),
            "color": dom.select(".info_table td")[11].text.strip(),
            "fuel": dom.select(".info_table td")[13].text.strip(),
            "foreclosure_mortgage": dom.select(".info_table td")[15].text.strip(),
            "sale_number": dom.select(".info_table td")[17].text.strip(),
            "accident_status": dom.select(".info_table td")[19].text.strip(),
            "car_model": dom.select(".info_table td")[21].text.strip(),
            "number": dom.select(".info_table td")[23].text.strip()}
```

```python
import sqlite3
```

```python
con = sqlite3.connect("car.db")
cur = con.cursor() # cursor? 

cur.executescript("""
    DROP TABLE IF EXISTS table1; 
    CREATE TABLE table1(
        id INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL,
        car TEXT NOT NULL,
        price TEXT NOT NULL,
        model_yr TEXT,
        fuel_efficiency TEXT,
        drive TEXT,
        engine_displacement TEXT,
        shift TEXT,
        color TEXT,
        fuel TEXT,
        foreclosure_mortgage TEXT,
        sale_number TEXT,
        accident_status TEXT,
        car_model TEXT,
        number TEXT
    );
""")
con.commit()
```

```python
while queue: # true
    baseURL = queue.pop(0) 
    seen.append(baseURL)
    content = parseContent(baseURL)
    cur.execute("""
        INSERT INTO table1 
        (car, price, model_yr, fuel_efficiency, drive, engine_displacement, shift, color, fuel, foreclosure_mortgage, 
        sale_number, accident_status, car_model, number)
        VALUES(?,?,?,?,?,?,?,?,?,?,?,?,?,?)
    """, [content["car"], content["price"], content["model_yr"], content["fuel_efficiency"], content["drive"],
         content["engine_diaplacement"], content["shift"], content["color"], content["fuel"], content["foreclosure_mortgage"],
         content["sale_number"], content["accident_status"], content["car_model"], content["number"]])
    con.commit()
    
```

```python
contents= list()
while queue:
    baseURL=queue.pop(0)
    seen.append(baseURL)
    contents.append(parseContent(baseURL))
```

```python
df = pd.DataFrame(contents)
```

```python
df
```

```python
df['price'] = df['price'].str.replace('만원','')
```

```python
df['engine_diaplacement'] = df['engine_diaplacement'].str.replace('CC','')
df['drive'] = df['drive'].str.replace('km','')
df['accident_status'] = df['accident_status'].str.replace('무','0')
df['accident_status'] = df['accident_status'].str.replace('유','1')

df['foreclosure_mortgage'] = df['foreclosure_mortgage'].str.replace('무','0')
df['foreclosure_mortgage'] = df['foreclosure_mortgage'].str.replace('유','1')
df['shift'] = df['shift'].str.replace('수동','0')
df['shift'] = df['shift'].str.replace('자동','1')
```

```python
# '가솔린', '디젤', 'LPG'
df['fuel'] = df['fuel'].str.replace('가솔린','1')
df['fuel'] = df['fuel'].str.replace('디젤','2')
df['fuel'] = df['fuel'].str.replace('LPG','3')
```

```python
df['engine_diaplacement'] = df['engine_diaplacement'].str.replace(',','')
df['drive'] = df['drive'].str.replace(',','')
df['price'] = df['price'].str.replace(',','')
```

```python
df['drive'] = df['drive'].astype('int')
df['engine_diaplacement'] = df['engine_diaplacement'].astype('int')
df['price'] = df['price'].astype('int')
```

```python
df['shift'] = df['shift'].astype('bytes')
df['foreclosure_mortgage'] = df['foreclosure_mortgage'].astype('bytes')
df['accident_status'] = df['accident_status'].astype('bytes')
```

```python
df['fuel'] = df['fuel'].astype('category')
```
